
# hist(x)

See `~matplotlib.axes.Axes.hist`.


```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

# plot:
fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.show()
```

```python
import matplotlib.pyplot as plt

# Exemple de données
valeurs = [10, 20, 30, 40, 50]

# Création du graphique
plt.plot(valeurs)

# Ajout de l'annotation pour la dernière valeur
derniere_valeur = valeurs[-1]
plt.annotate(f'Dernière valeur: {derniere_valeur}', xy=(len(valeurs) - 1, derniere_valeur),
             xytext=(len(valeurs) - 2, derniere_valeur + 10),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Affichage du graphique
plt.show()
```
