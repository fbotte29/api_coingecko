```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px

y = ['COOP_CWK_R1', 'COOP_ENER_R2', 'COOP_ENERC_MEZZ', 'COOP_ENERC_URSC', 'COOP_Pniche_R1', 'T EXT']

df = pd.read_csv("temperatures_COOP_Ete2020.csv")
fig = px.line(df, x='Time', y=y)

fig.add_hline(y=28, line_width=1, line_dash="dash", line_color="blue")
fig.add_hline(y=30, line_width=1, line_dash="dash", line_color="red", annotation_text="Seuil inconfort 28", 
              annotation_position="bottom right",
              annotation_font_size=15,
              annotation_font_color="blue")

              
fig.update_layout(xaxis_range=['2020-07-03','2020-10-03'],
                  title_text="Corrlation T EXT - TINT ETIC COOP")
#fig.update_xaxes(rangeslider_visible=True)
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=1, label="1d", step="day", stepmode="backward"),
            dict(count=1, label="1h", step="hour", stepmode="todate"),
            dict(step="all")
        ])
    )
)
fig.show()
```
